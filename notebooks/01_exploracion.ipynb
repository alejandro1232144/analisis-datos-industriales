# **Anlisis univariado**

Los histogramas muestran la distribucin de las principales variables del sistema. Las variables presentan rangos de operacin estables, lo que indica un comportamiento normal del sistema bajo condiciones estndar.

```python
# Seal de vibracin en el tiempo
df["timestamp"] = pd.to_datetime(df["timestamp"])

plt.figure(figsize=(12,4))
plt.plot(df["timestamp"], df["vibration"], label="Vibracin")
plt.xlabel("Tiempo")
plt.ylabel("Vibracin")
plt.title("Seal de vibracin en el tiempo")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Parmetros de ventaneo
fs = 1        # frecuencia de muestreo (1 muestra por unidad de tiempo)
w_sec = 20    # tamao de ventana en segundos
w_size = fs * w_sec

w_size
```

```python
def extraer_features(signal):
    return {
        "mean": signal.mean(),
        "std": signal.std(),
        "rms": np.sqrt(np.mean(signal**2)),
        "peak": signal.max() - signal.min()
    }
```

```python
features = []

for i in range(0, len(df) - w_size, w_size):
    ventana = df["vibration"].iloc[i:i+w_size]
    feats = extraer_features(ventana)
    feats["machine_failure"] = df["machine_failure"].iloc[i:i+w_size].max()
    features.append(feats)

df_features = pd.DataFrame(features)
df_features.head()
```

```python
df_features.info()
```

```python
baseline = df_features[df_features["machine_failure"] == 0]

umbral_rms = baseline["rms"].mean() + 2 * baseline["rms"].std()
umbral_rms
```

```python
df_features["anom"] = (df_features["rms"] > umbral_rms).astype(int)

df_features.head()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df_features["rms"], label="RMS")
plt.axhline(umbral_rms, color="red", linestyle="--", label="Umbral")
plt.xlabel("Ventana")
plt.ylabel("RMS")
plt.title("Deteccin de anomalas basada en RMS")
plt.legend()
plt.show()
```

```python
eventos = df_features[df_features["anom"] == 1]
eventos
```

## Deteccin de anomalas

Se defini un umbral basado en el comportamiento normal del RMS
(media + 2desviacin estndar). Bajo este criterio no se detectaron
ventanas fuera de rango, lo que indica que el sistema presenta un
funcionamiento estable durante el periodo analizado.
